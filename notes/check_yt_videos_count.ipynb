# 估算 YouTube 影片總數 (tubestats.org 方法概念驗證)

本筆記旨在實作 `tubestats.org` 背後的核心思想：透過**隨機抽樣 (Random Sampling)** 的統計方法，來估算整個 YouTube 平台上的影片總數量。

### 方法論

這個問題的核心挑戰在於，YouTube 的影片總數是一個巨大且不斷變動的數字，我們不可能窮舉所有影片來計數。因此，我們採用統計學的估算方法。

1.  **確定樣本空間**：
    *   一個標準的 YouTube 影片 ID 由 11 個字元組成。
    *   每個字元可以是大小寫字母、數字、底線 (`_`) 或減號 (`-`)，共 64 種可能性。
    *   因此，所有可能的影片 ID 總數是一個天文數字：$64^{11}$ (約 $7.3 \times 10^{19}$)。這就是我們的「樣本空間」。

2.  **進行隨機抽樣**：
    *   我們從這個巨大的樣本空間中隨機抽取一小部分樣本，例如，隨機生成 100 個符合格式的影片 ID。
    *   對於每一個隨機生成的 ID，我們需要一個方法來驗證它是否對應一個真實存在的影片。

3.  **計算命中率並估算總數**：
    *   計算在我們的樣本中，「命中」真實影片的比例 (命中率 = 命中數 / 總樣本數)。
    *   假設我們抽樣的命中率可以代表整個樣本空間的「影片密度」。
    *   **估算總數 = (總樣本空間大小) × (命中率)**

### 實作注意事項

*   **ID 驗證方法**：為了快速驗證 ID，我們將利用 YouTube 的 `oEmbed` 端點。向 `https://www.youtube.com/oembed?url=http://www.youtube.com/watch?v=VIDEO_ID` 發送請求，如果返回 `200 OK` 狀態碼，代表影片存在；如果返回 `404 Not Found`，則不存在。這比爬取整個影片頁面要高效得多。
*   **樣本大小**：本實作只會使用一個較小的樣本數 (例如 100 次) 來快速展示概念。在真實世界中，`tubestats.org` 會使用數百萬甚至數千萬的樣本來提高估算的準確性。樣本數越大，結果越接近真實值。
