```python
import numpy as np
from matplotlib import pyplot as plt
import torch

#离散动作
x = []
y = []
for i in np.linspace(0, 1, 100):
    prob = torch.FloatTensor([[i, 1 - i], [i, 1 - i]])

    entropy = prob * (prob + 1e-8).log()
    entropy = -entropy.sum(dim=1, keepdim=True).mean().item()

    x.append(i)
    y.append(entropy)

plt.figure(figsize=(4, 2))
plt.plot(x, y)
plt.show()
```

```python
#连续动作
x = []
y = []
for i in np.linspace(0.01, 2, 100):
    mu = torch.randn(2, 1)
    sigma = torch.FloatTensor([[i], [i]])

    dist = torch.distributions.Normal(mu, sigma)
    entropy = dist.entropy().mean().item()

    x.append(i)
    y.append(entropy)

plt.figure(figsize=(4, 2))
plt.plot(x, y)
plt.show()
```
